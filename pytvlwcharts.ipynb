```python
FifteenMinuteChart = Chart(
        layout=LayoutOptions(
            background_color='#222',
            text_color='#DDD',
            font_family="'Roboto', sans-serif",
        ),
        grid=GridOptions(
            horz_lines=GridLineOptions(
                color='#444',
                style=LineStyle.LINESTYLE_0,
                visible=False,
            ),
            vert_lines=GridLineOptions(
                color='#444',
                style=LineStyle.LINESTYLE_0,
                visible=False,
            ),        
        ),
        width = 1360,
        height = 400,
        time_scale=TimeScaleOptions(
            seconds_visible=True,
            time_visible=True,
            border_visible=True,
            border_color='#71649C',
        ),
        price_scale=PriceScaleOptions(
            border_visible=True,
            border_color='#71649C',
        )
    )
FifteenMinuteChart.mark_area(data=df15[['time','close']].rename(columns={"close": 'value'}),
                             scaleMargins={'top': 0.1,'bottom': 0.4},
                             topColor="rgba(56, 33, 110,0.6)",
                             bottomColor="rgba(56, 33, 110, 0.1)",
                             lineColor='transparent',
                             lineWidth=LineWidth.LINEWIDTH_1,
                             linestyle=LineStyle.LINESTYLE_0
                             )
FifteenMinuteChart.mark_candlestick(data=df15,scaleMargins={'top': 0.1,'bottom': 0.4},
        wickUpColor="rgb(54, 116, 217)",
        upColor="rgb(54, 116, 217)",
        wickDownColor="rgb(225, 50, 85)",
        downColor="rgb(225, 50, 85)",
        borderVisible=False,
        )
volume_df15['color'] = np.select([(bn_df_15['open'] < bn_df_15['close']), (bn_df_15['open'] > bn_df_15['close'])],["rgb(54, 116, 217)", "rgb(225, 50, 85)"])
FifteenMinuteChart.mark_histogram(data=volume_df15,
                                  priceFormat={'type': 'volume'},priceScaleId='',scaleMargins={'top': 0.7,'bottom': 0})
```
